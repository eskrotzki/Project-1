```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Import CSV datasets
gdp = "GDP_percap.csv"
life = "life_expec.csv"

gdp_data = pd.read_csv(gdp)
life_expec = pd.read_csv(life)

gdp_data_df = gdp_data.fillna(0)
life_expec_df = life_expec.fillna(0)
```

```python
life_expec_df = life_expec_df.set_index("Country Name")
```

```python
gdp_data_df = --gdp_data_df.set_index("Country Name")
```

```python
#Rounding dataframe to two decimal places 
life_expec_df = life_expec_df.round(decimals=2)
life_expec_df.head()
```

```python
#Rounding dataframe to two decimal places 
gdp_data_df = gdp_data_df.round(decimals=2)
gdp_data_df.head()
```

```python
# Making Scatter Plots
year_gdp = np.arange(1960,2020,1)
year_life = np.arange(1960,2019,1)
gdp_avg = gdp_data_df.mean(axis=0)
life_avg= life_expec_df.mean(axis=0)

fig1 = plt.scatter(year_gdp, gdp_avg, marker="o", facecolors="red", edgecolors="black", alpha=0.75)
plt.xlabel("Year")
plt.ylabel("Average GDP per Captia")

plt.show()
```

```python
fig2 = plt.scatter(year_life, life_avg, marker="o", facecolors="blue", edgecolors="black", alpha=0.75)
plt.xlabel("Year")
plt.ylabel("Average Life Expectancy")

plt.show()
```
